**Датасет - вино**
<br>Гауссов наивный байесовский классификатор

```python
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titles = ["Class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids",
           "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]  
data = pd.read_csv("wine.data", names=titles)
data["Class"] = data["Class"].astype("category")
data.sample(10)
```

```python
data.isna().sum()
```

```python
classes = data["Class"].unique()
classes_count = len(classes)
y = label_binarize(data["Class"], classes=classes)
X = data[["Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids",
           "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

```python
gauss = OneVsRestClassifier(GaussianNB())

gauss.fit(X_train, y_train)

y_pred = gauss.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

**Датасет - португальские школьники**
<br>Полиномиальный наивный байесовский классификатор

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
pd.options.display.max_columns = 999
data = pd.read_csv("student-mat.csv", sep=';')
data = data.astype({"school": "category"})
data.sample(10)
```

```python
data.isna().sum()
```

```python
sexBinary = {"M":0, "F":1}
data['sex_bin'] = data['sex'].map(sexBinary)

yesNo = {"yes":1, "no":0}
namesList = ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']

for i in namesList:
    data[i + '_bin'] = data[i].map(yesNo)
```

```python
X = data[['age', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'health', 'G1', 'G2', 'G3', 'sex_bin', 'schoolsup_bin', 'famsup_bin', 'paid_bin', 'activities_bin', 'nursery_bin', 'higher_bin', 'internet_bin']]
y = data['school'].copy().cat.codes

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
multi = MultinomialNB()
y_pred = multi.fit(X_train, y_train).predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

**Датасет - вино**
<br>Метрический алгоритм k-ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
columns = ["Class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids",
           "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]  
wine = pd.read_csv('wine.data', names=columns)
wine['Class'] = wine['Class'].astype('category')
wine.sample(10)
```

```python
wine.isna().sum()
```

```python
X = wine.iloc[:, 1:13]
y = wine.iloc[:, 0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24, shuffle=True)
knc = KNeighborsClassifier(n_neighbors=5)
knc.fit(X_train, y_train)
predict = knc.predict(X_test)

accuracy_score(predict, y_test)
```

```python
cv = GridSearchCV(KNeighborsClassifier(), {'n_neighbors': range(1, 26)})
cv.fit(X, y)
cv.best_params_
```
