**Датасет - Португальские школьники**
<br>Бинарная классификация

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble  import RandomForestClassifier
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score

%matplotlib inline
```

```python
pd.options.display.max_columns = 999
data = pd.read_csv("student-mat.csv", sep=';')
data = data.astype({"school": "category"})
data.sample(10)
```

```python
data.isna().sum()
```

```python
sexBinary = {"M":0, "F":1}
data['sex_bin'] = data['sex'].map(sexBinary)

yesNo = {"yes":1, "no":0}
namesList = ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']

for i in namesList:
    data[i + '_bin'] = data[i].map(yesNo)

data.sample(10)
```

```python
X = data[['age', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'health', 'G1', 'G2', 'G3', 'sex_bin', 'schoolsup_bin', 'famsup_bin', 'paid_bin', 'activities_bin', 'nursery_bin', 'higher_bin', 'internet_bin']]
y = data['school'].copy().cat.codes

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rfc = RandomForestClassifier(criterion='entropy', n_estimators=700, min_samples_split=10, min_samples_leaf=1, max_features='auto', oob_score=True, random_state=1, n_jobs=-1)

rfc.fit(X_train, np.ravel(y_train))
```

```python
predict = rfc.predict_proba(X_test)
predict = predict[:, 1]

auc = roc_auc_score(y_test, predict)
auc
```

```python
fpr, tpr, _ = roc_curve(y_test, predict)
plt.plot([0,1], [0,1], linestyle='-')
plt.plot(fpr, tpr, marker='.')
plt.show()
```

**Датасет - вино**
<br>Многоклассовая классификация

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
columns = ["Class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids",
           "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]  
wine = pd.read_csv('wine.data', names=columns)
wine['Class'] = wine['Class'].astype('category')
wine.sample(10)
```

```python
wine.isna().sum()
```

```python
X = wine.iloc[:, 1:13]
y = wine.iloc[:, 0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=24, shuffle=True)
knc = KNeighborsClassifier(n_neighbors=1)
knc.fit(X_train, y_train)
predict = knc.predict(X_test)

accuracy_score(predict, y_test)
```

**Датасет - пожары**
<br>Регрессия

```python
from sklearn.linear_model import LinearRegression
```

```python
fires = pd.read_csv('forestfires.csv')
fires.sample(10)
```

```python
fires.isna().sum()
```

```python
y = fires[['temp']]
X = fires[['FFMC', 'DMC', 'DC', 'ISI', 'RH', 'wind', 'rain']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=45)
y_pred = LinearRegression().fit(X_train, y_train).predict(X_test)
```

```python
miss = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(miss)
print(r2)
```

```python
_, view = plt.subplots()
view.scatter(y_test, y_pred, s = 5, color = "r", alpha = 0.75)
display()
```
