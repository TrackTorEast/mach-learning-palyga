```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import roc_auc_score, accuracy_score, roc_curve

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv("student-mat.csv", sep=';')
pd.options.display.max_columns = 999
data.sample(40)
```

```python
data = data[['sex', 'age', 'studytime', 'G1']]
data['sex'] = data['sex'].replace({'M': 1, 'F': 0})
data.sample(10)
```

```python
X = data.iloc[:, 0:3]
y = data['G1']
y[y<=10] = 0
y[y>10] = 1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=32, shuffle=True)
```

```python
clf = DecisionTreeClassifier(random_state=51)
clf.fit(X_train, y_train)
list(zip(X.columns, clf.feature_importances_))
```

```python
y_pred = clf.predict(X_test)
roc_auc_score(y_test, y_pred)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.plot([0,1], [0,1], linestyle='-')
plt.plot(fpr, tpr, marker='.')
plt.show()
```
